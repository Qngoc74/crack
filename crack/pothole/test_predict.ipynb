# predict yolo 1 ảnh

```python
from ultralytics import YOLO
from IPython.display import display, Image

# ==============================
# ĐƯỜNG DẪN
# ==============================
model_path = "best.pt"              # đổi lại nếu file nằm chỗ khác
image_path = "test.jpg"             # ảnh cần predict
save_dir = "runs/predict_image"     # thư mục lưu kết quả

# ==============================
# LOAD MODEL
# ==============================
model = YOLO(model_path)

# ==============================
# PREDICT ẢNH
# ==============================
results = model.predict(
    source=image_path,
    save=True,
    project=save_dir,
    name="exp",
    conf=0.25,      # ngưỡng confidence
    iou=0.45,       # ngưỡng IoU cho NMS
    show=False
)

# ==============================
# HIỂN THỊ ẢNH KẾT QUẢ
# ==============================
result_image_path = f"{save_dir}/exp/{image_path.split('/')[-1]}"
display(Image(filename=result_image_path))

# ==============================
# IN THÔNG TIN BOX
# ==============================
for r in results:
    print("Classes:", r.boxes.cls.cpu().numpy() if r.boxes is not None else [])
    print("Confidences:", r.boxes.conf.cpu().numpy() if r.boxes is not None else [])
    print("Boxes:", r.boxes.xyxy.cpu().numpy() if r.boxes is not None else [])
```

# predict 1 video

```python
from ultralytics import YOLO
from IPython.display import Video, display
import os

# ==============================
# ĐƯỜNG DẪN
# ==============================
model_path = "best.pt"               # đổi lại nếu file nằm chỗ khác
video_path = "test_video.mp4"        # video cần predict
save_dir = "runs/predict_video"      # thư mục lưu kết quả

# ==============================
# LOAD MODEL
# ==============================
model = YOLO(model_path)

# ==============================
# PREDICT VIDEO
# ==============================
results = model.predict(
    source=video_path,
    save=True,
    project=save_dir,
    name="exp",
    conf=0.25,
    iou=0.45,
    show=False
)

# ==============================
# TÌM FILE VIDEO OUTPUT
# ==============================
output_video_path = f"{save_dir}/exp/{os.path.basename(video_path)}"
print("Video đã lưu tại:", output_video_path)

# ==============================
# HIỂN THỊ VIDEO
# ==============================
display(Video(output_video_path, embed=True))
```
